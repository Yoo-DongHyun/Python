```python
_type = input('형식: (day or time): ')
stock_code = input("종목 코드: ")
page_num = int(input("페이지 (-1은 끝까지)"))
now = datetime.now()


url = 'https://finance.naver.com/item/site_{0}.nhn'.format(_type)
params = {
    'code': stock_code,
    'thistime' : now.strftime('%Y%m%d%H%M%S')
}
if(page_num==-1):
    page_num = 1000000000

result_list = []
for i in range(1, page_num+1):
    params['page'] = i
    resp = requests.get(url, params=params)
    
    soup = BeautifulSoup(resp.content)
    stock_table = soup.find('table',class_='type2')
    
    tr_tags = stock_table.find_all('tr')
    
   
        
    for tr_tag in tr_tags:
        td_tag = tr_tag.find_all('td')
        if len(td_tag)<7:
            continue
        result_list.append({
            'time' : td_tag[0].text,
            'closing_price' : td_tag[1].text,
            'day_before' : td_tag[2].text.strip(),
            'sell' : td_tag[3].text,
            'buy' : td_tag[4].text,
            'volume' : td_tag[5].text,
            'variation' : td_tag[6].text
        }) 
        if len(tr_tags)<10:
            break
            
```
