```python
resp = requests.get('https://search.naver.com/search.naver', params={'where' : 'news', 'query': '펭수'})
```

```python
resp
```

```python
resp.text
```

```python
soup = BeautifulSoup(resp.content)
```

```python
soup
```

```python
peng_news_list = soup.find('ul', class_="type01")
```

```python
li_tags = peng_news_list.find_all('li',recursive=False)
len(li_tags)
```

```python
peng_news_titles = peng_news_list.find_all('a', class_='_sp_each_title' )
```

```python
peng_news_titles
```

```python
news_list=[]

for news in peng_news_titles:
    title = news.text
    link = news.get('href')
    news_list.append({
        'title' : title,
        'link' : link
    })
news_list
```

```python
news_list=[]
for li_tag in li_tags:
    a_tag = li_tag.find('a', class_ = '_sp_each_title')
    title= a_tag.title
    link = a_tag.get('href')
    
    dd_tag=li_tag.find_all('dd')[1]
    desc = dd_tag.text

    news_list.append({
        'title': title,
        'link' : link,
        'desc' : desc
    })
news_list
```
